```python
from sklearn.linear_model import LogisticRegression
```

```python
params={'class_weight':['balanced',None],
        'penalty':['l1','l2'],
        'C':np.linspace(0.01,1000,10)}
```

```python
x_train=cv_train.drop('V86',axis=1)
y_train=cv_train['V86']
```

```python
 y_train.dtype
```

```python
model=LogisticRegression(fit_intercept=True)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search=GridSearchCV(model,param_grid=params,cv=10,scoring="roc_auc")
```

```python
grid_search.fit(x_train,y_train)
```

```python
fbetas=[]
```

```python
for cutoff in cutoffs:
    
    predicted=(train_score>cutoff).astype(int)
    
    fbetas.append(fbeta_score(y_train,predicted,2))
    
```

```python
my_cutoff=cutoffs[fbetas==max(fbetas)]
```

```python
predictions=(grid_search.predict_proba(cv_test)[:,1]>my_cutoff).astype(int)
```
